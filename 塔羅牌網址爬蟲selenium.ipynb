```python
#抓牌的網址
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import openpyxl as opx
wb = opx.Workbook()
ws = wb.active
ws["A1"] = "牌名"
ws["B1"] = "網址"

url = "https://www.tarnote.com/blog/你該知道的78張塔羅牌義與塔羅占卜知識大全持續更新中"

option = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values":{"notifications":2}}
option.add_experimental_option("prefs",prefs)
#瀏覽器路徑
driver = webdriver.Chrome("C:\\Users\\htuvm\\python\\chrome動態爬蟲工具\\chromedriver.exe",options=option)
driver.get(url)
data=[]
num = 2
while num<84:
    try:
        try:
            name = WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.XPATH,"/html/body/div[5]/div[1]/p["+str(num)+"]/a/u"))).text
            URL = WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.XPATH,"/html/body/div[5]/div[1]/p["+str(num)+"]/a"))).get_attribute("href")
        except:
            name = WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.XPATH,"/html/body/div[5]/div[1]/p["+str(num)+"]/u/a"))).text
            URL = WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.XPATH,"/html/body/div[5]/div[1]/p["+str(num)+"]/u/a"))).get_attribute("href")
    except:
        name = None
        URL = None
    num+=1
    ws.append([name,URL])
    wb.save("塔羅牌網站.xlsx")
    print("牌名:",name)
    print("網址:",URL)
```

```python
#導入excel網址
import pandas
import json
data = []
with open("塔羅牌網站.xlsx","rb") as f:
    pd = pandas.read_excel(f)
for url in pd["網址"]:
    data.append(url)
print(data)
```

```python
#抓圖片的位置
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import openpyxl as opx
wb = opx.Workbook()
ws = wb.active

ws["A1"] = "網址"


option = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values":{"notifications":2}}
option.add_experimental_option("prefs",prefs)
#瀏覽器路徑
driver = webdriver.Chrome("C:\\Users\\htuvm\\python\\chrome動態爬蟲工具\\chromedriver.exe",options=option)
num = 1
for x in data:
    driver.get(x)
    jpg = WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CSS_SELECTOR,'img[style]'))).get_attribute("src")
    ws.append([jpg])
    wb.save("塔羅牌圖片.xlsx")
    print("網址:",jpg,num)
    time.sleep(1)
    num+=1
```

```python
#抓牌義
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import openpyxl as opx
wb = opx.Workbook()
ws = wb.active

ws["B1"] = ""


option = webdriver.ChromeOptions()
prefs = {"profile.default_content_setting_values":{"notifications":2}}
option.add_experimental_option("prefs",prefs)
#瀏覽器路徑
driver = webdriver.Chrome("C:\\Users\\htuvm\\python\\chrome動態爬蟲工具\\chromedriver.exe",options=option)
num = 1
for x in data:
    driver.get(x)
    ele = WebDriverWait(driver,10).until(ec.element_to_be_clickable((By.CSS_SELECTOR,'div[class="ac-text"]'))).text
    ws.append([ele])
    wb.save("牌義.xlsx")
    print("牌義:",ele,num)
    time.sleep(1)
    num+=1
```
